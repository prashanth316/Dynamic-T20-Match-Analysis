```python
import pandas as pd

count = 1  # Initialize count variable
delivery_df = pd.DataFrame()  # Create an empty DataFrame to store delivery data

# Iterate over rows in the 'matches' DataFrame
for index, row in matches.iterrows():
    # Skip specific matches based on their index (count)
    if count in [75, 108, 150, 180, 268, 360, 443, 458, 584, 748, 982, 1052, 1111, 1226, 1345]:
        count += 1
        continue
    count += 1  # Increment count
    
    # Initialize lists to store data for each delivery
    ball_of_match = []
    batsman = []
    bowler = []
    runs = []
    player_of_dismissed = []
    teams = []
    batting_team = []
    match_id = []
    city = []
    venue = []
    winner = []
    
    # Iterate over deliveries in the first innings of the match
    for ball in row['innings'][0]['1st innings']['deliveries']:
        for key in ball.keys():  # Iterate over keys in each delivery
            # Append data to lists
            match_id.append(count)
            batting_team.append(row['innings'][0]['1st innings']['team'])
            teams.append(row['info.teams'])
            ball_of_match.append(key)
            batsman.append(ball[key]['batsman'])
            bowler.append(ball[key]['bowler'])
            runs.append(ball[key]['runs']['total'])
            city.append(row['info.city'])
            winner.append(row['info.outcome.winner'])
            venue.append(row['info.venue'])
            try:
                player_of_dismissed.append(ball[key]['wicket']['player_out'])
            except:
                player_of_dismissed.append('0')
    
    # Create a DataFrame for the current match's deliveries
    loop_df = pd.DataFrame({
        'match_id': match_id,
        'teams': teams,
        'batting_team': batting_team,
        'ball': ball_of_match,
        'batsman': batsman,
        'bowler': bowler,
        'runs': runs,
        'player_dismissed': player_of_dismissed,
        'city': city,
        'venue': venue,
        'winner' : winner
    })
    
    # Append the DataFrame for the current match to the delivery_df DataFrame
    delivery_df = pd.concat([delivery_df, loop_df])

delivery_df
```
